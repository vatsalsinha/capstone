```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./fully_processed_college_db.csv')
df.shape
```

```python
df.iloc[0:5,16:]
```

```python
#User input variables, defined manually here
u_placements = 3
u_food = 4
u_safety = 4
u_hostels = 2
u_on_campus_life = 2
u_scholarships = 2
u_infrastructure = 1
u_research = 2
u_study_abroad = 3
u_fests = 3
u_clubs = 3
u_diversity = 4
u_sports_gym = 2
u_alumni_network = 3
u_transportation_facilities = 4

u_state = 'Karnataka'
```

```python
df = df.loc[df['state'] == u_state]
df.iloc[0:10, 0:16]
```

```python
df.shape
```

```python
df['Score'] = ''
```

```python
df.shape
```

```python
from scipy import spatial
w1,w2,w3 = 3,2,1
def score(u,c):
    sc = w1*(1 - spatial.distance.cosine(u[0:4],c[0:4])) +  w2*(1 - spatial.distance.cosine(u[4:13],c[4:13])) + w3*(1 - spatial.distance.cosine(u[13:],c[13:]))
    p = (sc / 6) * 100
    return p

user = [u_placements,u_food,u_safety,u_hostels,u_on_campus_life,u_scholarships,u_infrastructure,u_research,
            u_study_abroad,u_fests,u_clubs,u_diversity,u_sports_gym,u_alumni_network,u_transportation_facilities]
for ind in df.index:
    college = [df['r_placements'][ind], df['r_food'][ind], df['r_safety'][ind], df['r_hostels'][ind], 
               df['r_on_campus_life'][ind], df['r_scholarships'][ind], df['r_infrastructure'][ind], df['r_research'][ind],  
              df['r_study_abroad'][ind], df['r_fests'][ind], df['r_clubs'][ind], df['r_diversity'][ind], 
               df['r_sports_gym'][ind], df['r_alumni_network'][ind], df['r_transportation_facilities'][ind]]
    df['Score'][ind] = score(user,college)

    
```

```python
df.iloc[0:10,31:]
```

```python
df = df.sort_values(by="Score",ascending=False)
df.iloc[0:10,31:]
```
